```python
match_df = matches.merge(total_score_df[['match_id', 'total_runs']], left_on='id', right_on= 'match_id')

# merged total score df with match df and put total score of first inning of each match in match_df
```

```python
#only keep those team which are currently playing

teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bengaluru',
    'Kolkata Knight Riders',
    'Punjab Kings',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Gujarat Titans',
    'Lucknow Super Giants'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Royal Challengers Banglaore', 'Royal Challengers Bengaluru')
match_df['team2'] = match_df['team2'].str.replace('Royal Challengers Banglaore', 'Royal Challengers Bengaluru')

match_df['team1'] = match_df['team1'].str.replace('Kings XI Punjab', 'Punjab Kings')
match_df['team2'] = match_df['team2'].str.replace('Kings XI Punjab', 'Punjab Kings')
```

```python
match_df = match_df[
    (match_df['team1'].isin(teams)) &
    (match_df['team2'].isin(teams))
]
```

```python
match_df.shape
```

```python
match_df['team1'].unique()
```

```python
match_df =match_df[match_df['method']!='D/L']
```

```python
match_df.shape
```

```python
match_df = match_df[['match_id', 'city', 'winner', 'total_runs']]
```

```python
match_df.tail()
```

```python
deliveries_df = match_df.merge(deliveries, on='match_id')
```

```python
deliveries_df.tail()
```

```python
deliveries_df = deliveries_df[deliveries_df['inning'] == 2]
```

```python
deliveries_df.shape
```

```python
deliveries_df.sample(5)
```

```python
deliveries_df['current_score'] = deliveries_df.groupby(['match_id'])['total_runs_y'].cumsum()
```

```python
deliveries_df['runs_left'] = deliveries_df['total_runs_x'] +1 - deliveries_df['current_score']
```

```python
deliveries_df.tail()
```

```python
deliveries_df['balls_left'] = 126 - (deliveries_df['over']*6 + deliveries_df['ball'])
```

```python
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].fillna('0')
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].apply(lambda x: x if x == '0' else '1')
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].astype('int')
deliveries_df['wickets_fall'] = deliveries_df.groupby(['match_id'])['player_dismissed'].cumsum()
deliveries_df['wickets_left'] = 10 - deliveries_df['wickets_fall']
```

```python
deliveries_df.tail()
```

```python
deliveries_df['CRR'] = (deliveries_df['current_score']*6)/(120-deliveries_df['balls_left'])
deliveries_df['RRR'] = (deliveries_df['runs_left']*6)/deliveries_df['balls_left']
```

```python
#we are only using second inning data, so if winner is same as abatting team, we encodde as 1, else 0

def result(row):
  return 1 if row['batting_team'] == row['winner'] else 0
```

```python
deliveries_df['result'] = deliveries_df.apply(result, axis=1)
```

```python
final_df= deliveries_df[['batting_team', 'bowling_team', 'city','runs_left', 'balls_left', 'wickets_left', 'total_runs_x', 'CRR', 'RRR', 'result']]
```

```python
final_df.tail()
```

```python
#suffling to minimize baisness

final_df = final_df.sample(final_df.shape[0])
final_df = final_df[final_df['balls_left']!=120]
```

```python
final_df.shape
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
x =final_df.drop(columns=['result'])
y =final_df['result']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
#
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
]
)
```

```python
final_df.describe()
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
pipe2 = Pipeline(steps=[
    ('step1', trf),
    ('step2', RandomForestClassifier())
]
)
pipe2.fit(x_train, y_train)
y_pred2 =pipe2.predict(x_test)
accuracy_score(y_test, y_pred2)
#
```
